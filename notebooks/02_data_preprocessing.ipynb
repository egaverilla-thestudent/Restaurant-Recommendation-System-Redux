```python
import pandas as pd

business = pd.read_json("/Users/edgardomosesezekielaverilla/Restaurant-Recommendation-System-Redux/yelp_dataset/yelp_academic_dataset_business.json", lines=True)

print(f"Rows: {business.shape[0]}")
print(f"Columns: {business.shape[1]}")
```

```python
restaurants = business[
    business["categories"].str.contains(
        "Restaurants",
        case=False,
        na=False
    )
].copy()
```

```python
restaurants.shape
```

```python
columns = ["business_id", "name", "city", "state", "postal_code", "latitude", "longitude", "stars", "review_count", "categories", "attributes", "hours", "is_open"]

restaurants = restaurants[columns]

```

Missing Value Handling


```python
restaurants = restaurants.fillna({"attributes": "Uknown", "hours": "Uknown"})
```
